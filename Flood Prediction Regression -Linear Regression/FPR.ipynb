```python
X = df.loc[:, df.columns != "FloodProbability"]
y = df['FloodProbability']
```

```python
X.columns
```

```python
y.unique()
```

```python
#spliting the dataset 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# model buidling 
from sklearn.linear_model import LinearRegression
Regression=LinearRegression()
Regression.fit(X_train,Y_train)
```

```python
Y_predict=Regression.predict(X_test)
```

```python
print("Intercept ",Regression.intercept_)
print("the scope of weight ",Regression.coef_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("Mean absoute error ",mean_absolute_error(Y_test,Y_predict))
print("Mean sqaure error ",mean_squared_error(Y_test,Y_predict))
print("R2 score",r2_score(Y_test,Y_predict))
```

```python
Y_predict_test=Regression.predict(df_test)
```
